```python
import pandas as pd

train = pd.read_csv("../data/train.csv")
train.head()
```

```python
train.shape
train.info()
train.describe()
```

There are 79 features in total. The data contains both numeric and categorical variables. Several columns contain missing values, especially poolQC, Alley, Fence, etc.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(train['SalePrice'], kde=True)
plt.title("Sale Price Distribution")
```

SalePrice is right-skewed.

```python
sns.scatterplot(x=train["GrLivArea"], y=train["SalePrice"])
plt.title("GrLivArea vs SalePrice")
```

Dependency: SalePrice depends on GrLivArea.

Larger living areas are associated with higher sale prices.

```python
sns.boxplot(x=train["OverallQual"], y=train["SalePrice"])
plt.title("SalePrice vs Overall Quality")
```

Dependency: SalePrice depends on OverallQual.

OverallQual is the strongest predictor of SalePrice.
As the overall quality increases, the price rises very sharply.
Values 8-10 correspond to high-end or "luxury" houses.

```python
corr = train.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    cmap="RdBu_r",
    square=True,
    linewidths=0.5
)

plt.title("Correlation Heatmap (numeric features)")
plt.tight_layout()
```

Dependency: Features depend on each other.

The color of each cell represents the strength and direction of the linear relationship between two variables.
Looking at the SalePrice row/column, we see that OverallQual, GrLivArea, GarageCars GarageArea, TotalBsmtSF and 1stFlrSF show the strongest positive correlations with the target.
The heatmap also reveals groups of highly correlated predictors (e.g. GarageCars and GarageArea, TotalBsmtSF and 1stFlrSF).

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=train, x="Neighborhood", y="SalePrice")
plt.xticks(rotation=90)
plt.show()
```

Dependency: SalePrice depends on Neighborhood.

High-price neighborhoods include StoneBr, NridgHt and NoRidge.
Lower-price neighborhoods include IDOTRR and MeadowV.
